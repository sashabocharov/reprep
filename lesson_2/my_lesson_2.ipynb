```python
df = df.rename(columns={'Номер': 'number', 
                        'Дата создания': 'create_date',
                        'Дата оплаты': 'payment_date',
                        'Title': 'title',
                        'Статус': 'status',
                        'Заработано': 'money',
                        'Город': 'city',
                        'Платежная система': 'payment_system'
                        }
                )
```

```python
df
```

```python
money_by_city = df\
        .query('status == "Завершен"')\
        .groupby(['title', 'city'], as_index=False)\
        .aggregate({'money': 'sum', 'number': 'count'})\
        .sort_values('money', ascending=False)\
        .rename(columns={'number': 'success_number'})\
        .reset_index(drop=True)
```

```python
taxi = pd.read_csv('./2_taxi_nyc.csv')
```

```python
taxi
```

```python
for i in taxi.columns:
    taxi.rename(columns={i: i.replace(' ', '_')}, inplace=True)
```

```python
taxi
```

```python
taxi['borough'].value_counts()
```

```python
taxi\
    .groupby('borough', as_index=False)\
    .pickups.sum()\
    .sort_values('pickups', ascending=False)
```

```python
min_pickups = taxi\
    .groupby('borough')\
    .aggregate({'pickups':'sum'})\
    .pickups.idxmin()
```

```python
min_pickups
```

```python
avg_pickups = taxi\
    .groupby(['borough', 'hday'], as_index=False)\
    .pickups.mean()\
    .sort_values(['borough', 'hday'], ascending=[True, False])
avg_pickups
```

```python
pickups_by_mon_bor = taxi\
    .groupby(['pickup_month', 'borough'], as_index=False)\
    .agg({'pickups': 'sum'})\
    .sort_values('pickups', ascending=False)
pickups_by_mon_bor
```
