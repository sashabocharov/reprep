```python
import pandas as pd
```

```python
bookings = pd.read_csv(r'./2_bookings.csv', sep=';')
```

```python
bookings.columns.str.lower()
```

```python
for i in bookings.columns:
    bookings.rename(columns={i: i.replace(' ', '_')}, inplace= True)
bookings.columns = bookings.columns.str.lower()
```

```python
bookings
```

```python
success_bookings = bookings\
                    .query('is_canceled == 0')\
                    .groupby('country')\
                    .size()\
                    .sort_values(ascending=False)\
                    .head(5)
success_bookings
```

```python
avg_nights = bookings\
            .groupby('hotel', as_index=False)\
            .agg({'stays_total_nights': 'mean'})\
            .round(2)\
            .sort_values('stays_total_nights', ascending=False)
avg_nights
```

```python
diff_suits = bookings.query('assigned_room_type != reserved_room_type').shape[0]
diff_suits
```

```python
arrival_date = bookings\
                .query("arrival_date_year in [2016, 2017]")\
                .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False)\
                .arrival_date_year.value_counts()\
                .sort_values(by='count',ascending=False)
arrival_date
```

```python
otmena = bookings\
        .query('hotel == "City Hotel" & is_canceled == 1')\
        .groupby('arrival_date_year', as_index=False)\
        .arrival_date_month.value_counts()\
        .sort_values(['arrival_date_year', 'count'], ascending=[False, False])
otmena
```

```python
bookings['total_kids'] = bookings.children + bookings.babies
```

```python
most_child = bookings\
            .groupby('hotel', as_index=False)\
            .total_kids.mean().round(2).max()
most_child
```

```python
kids = bookings.groupby('has_kids', as_index=False)['is_canceled'].value_counts(normalize=True)
kids
```
