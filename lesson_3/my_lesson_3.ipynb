```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

%matplotlib inline
```

```python
df = pd.read_csv('lesson_3_data_1_.csv', encoding='windows-1252')
```

```python
df
```

```python
user_df = df[['tc', 'art_sp']]
```

```python
user_df
```

```python
user_df = user_df.rename(columns={'tc': 'user_id', 'art_sp': 'brand_info'})
```

```python
user_df
```

```python
# user_df.brand_info.apply(lambda x: x.split(' ')[-1])
```

```python
def split_brand(brand_name):
    return brand_name.split(' ')[-1]
user_df['brand_name'] = user_df.brand_info.apply(split_brand)
user_df
```

```python
user_purchases = user_df\
                .groupby('user_id', as_index=False)\
                .agg({'brand_name': 'count'})\
                .rename(columns={'brand_name': 'purchases'})\
                .query('purchases >= 5')
user_purchases
```

```python
users_unique_brands = user_df\
                    .groupby('user_id', as_index=False)\
                    .agg({'brand_name': pd.Series.nunique})\
                    .rename(columns={'brand_name': 'unique_brands'})
users_unique_brands
```

```python
lovely_brand_purchases_df = user_df\
    .groupby(['user_id', 'brand_name'], as_index=False)\
    .agg({'brand_info': 'count'})\
    .sort_values(['user_id', 'brand_info'], ascending=[False, False])\
    .groupby('user_id', as_index=False)\
    .head(1)\
    .rename(columns={'brand_name': 'lovely_brand', 'brand_info': 'lovely_brand_purchases'})
lovely_brand_purchases_df
```

```python
loyality_df = user_purchases\
            .merge(users_unique_brands, on='user_id')\
            .merge(lovely_brand_purchases_df, on='user_id')
loyality_df
```

```python
loyal_users = loyality_df[loyality_df['unique_brands'] == 1]
loyal_users
```

```python
loyality_df['loyality_score'] = (loyality_df.lovely_brand_purchases / loyality_df.purchases * 100).round()
loyality_df
```

```python
ax = sns.displot(loyality_df.loyality_score)
```

```python
brand_loyality = loyality_df\
                .groupby('lovely_brand', as_index=False)\
                .agg({'loyality_score': 'median', 'user_id': 'count'})
brand_loyality
```

```python
dx = sns.barplot(x = 'lovely_brand', y = 'user_id', data = brand_loyality)
```
